## 0. 경로설정

```python
path = "./data"
file_list = list(os.listdir(path))

print ("file_list: {}".format(file_list))
```

## 1. Naver_Blog

```python
# A _ blog_contents
blog_contents = pd.read_csv(path+'/'+file_list[5], encoding='utf-8')
# blog_contents.iloc[:1]

# 두개의 열 추출 
A_a = blog_contents[['posting_date','title']]
A_b = blog_contents[['posting_date','contents']]

# 컬럼명 변경 -> 컬럼명을 바꿔서, CONCAT하기 위해
A_a.rename(columns = {'title' : 'content'}, inplace = True)
A_b.rename(columns = {'contents' : 'content'}, inplace = True)

# 한 줄로 CONCAT
A = pd.concat([A_a,A_b],axis=0).reset_index(drop = True)
A.rename(columns = {'posting_date' : 'date'}, inplace = True)
A.head()
```

```python
# A.info()
```

```python
# B_ blog_review

blog_comments = pd.read_csv(path+'/'+file_list[1], encoding='utf-8')
# blog_comments.iloc[:1]
B = blog_comments[['comment_date','comment']]

# 컬럼명 변경 -> 컬럼명을 바꿔서, CONCAT하기 위해
B.rename(columns = {'comment' : 'content'}, inplace = True)
B.rename(columns = {'comment_date' : 'date'}, inplace = True)

B.head()
```

```python
# naver_blog 내용, 댓글 합치기 
Naver_Blog = pd.concat([A,B],axis = 0).reset_index(drop = True)

# 플랫폼 넘버 삽입
Naver_Blog.insert(0,'platform_num',1)

# date 부분 년-월-일로 정렬 
date = []
for i in Naver_Blog['date']:
    j = str(i).split()[0]
    date.append(j)
    
Naver_Blog['date'] = date
Naver_Blog
```

## 2. Naver_Kin

```python
# C_kin_content

kin_contents = pd.read_csv(path+'/'+file_list[3], encoding='utf-8')
# kin_contents.iloc[:3]

# 두개의 열 추출 
C_a = kin_contents[['date','content']]
C_b = kin_contents[['date','question']]

# 컬럼명 변경 -> 컬럼명을 바꿔서, CONCAT하기 위해
C_b.rename(columns = {'question' : 'content'}, inplace = True)

# 한 줄로 CONCAT
C = pd.concat([C_a,C_b],axis=0).reset_index(drop = True)
C.head()

# 지식인으로 이름 변경
Naver_Kin = C.copy()

# 플랫폼 넘버 삽입
Naver_Kin.insert(0,'platform_num',2)

# date 부분 년-월-일로 정렬 
date = []
for i in Naver_Kin['date']:
    j = str(i).replace('.','-')
    j = j[:-1]
    date.append(j)
Naver_Kin['date'] = date

Naver_Kin
```
