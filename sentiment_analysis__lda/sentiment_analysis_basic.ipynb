```python
# 패키지 정리 

import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
# Google colab: 데이터 경로 설정
path = '/content/gdrive/My Drive/endustrial_engineering_conference_project'

# 파일 읽기
data = pd.read_csv(path + '/pre_finished_data/final_data.csv', index_col = 0)
data.tail()
```

```python
# 1. 한글만 추출 

# 2. 오타교정

# 3. 맞춤법 교정

# 4. 문장 띄어쓰기 수행 

# 5. pororo 감성 점수 추출 

# 6. 최종 추출 결과를 가지고 2점이하는 negative, 4점이상은 positve로 분류
```

```python
# 1. 한글만 추출

# 한글만 추출 -> 특수문자 제거
data['content'] = data['content'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
# print(data['content'].iloc[:2])
```

```python
# 2. 오타 교정 
```

```python
# 3. 문장 띄어쓰기 수행
# https://docs.likejazz.com/kss/

# ! pip install kss
import kss

# 리스트 타입으로 변경 
sents = [sentence for sentence in data['content']]
sents[0]

# 한 문장씩 추출 
final_sent = []
for index,sent in enumerate(sents) : # 일부만 테스트 
    try:
      for each_sent in kss.split_sentences(sent):
        final_sent.append(each_sent)
    except AssertionError:
      print(index,sent)
    
final_sent
```

```python
# 4. 맞춤법교정
!pip install git+https://github.com/ssut/py-hanspell.git
from hanspell import spell_checker
final_sent_2 = []
for sent in final_sent : 
    result = spell_checker.check(sent).checked
    final_sent_2.append(result)
final_sent_2
```

```python
# 5. pororo 감성점수 추출

review = Pororo(task='review', lang = 'ko')
# review('그냥저냥 다른데랑 똑같습니다.')

review_rate = []
for index,sent in enumerate(final_sent_2):
    try:
      review_rate.append(review(sent))
    except ValueError :
      print(index,sent)

review_rate
```

```python
# 데이터 프레임 합치기 
sentiment_dict = {
    'each_sent':final_sent,
    'correct_sent':final_sent_2,
    'sentiment ratio':review_rate
}
```

```python
sentiment_df = pd.DataFrame(sentiment_dict)
print(sentiment_df)
```

```python
filename = 'sentiment_df'
sentiment_df.to_csv(path + '/network_anaysis/co_occurrence_matrix/' + filename + '.csv', sep = ',',encoding='utf-8-sig')
```
